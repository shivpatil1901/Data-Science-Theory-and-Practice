### Q2

#### Bracketing method

```python

def J(w):
    if w==0:
        return float('inf')
    else:
        return w**2 +(54/w)

# Using bracketing method
w1=a=0
b=5
n=10
delw=(b-a)/n
print('The interval is:',delw)
# def bracketing()
w2=w1+delw
w3=w2+delw
for w1 in range(5):
    w1=w2
    w2=w3
    w3=w3+delw
    
    if J(w1)>=J(w2) and J(w2)<=J(w3):
        print('Minima lies between:',w1,'and',w3)
        break
    if w3>=5:
        break
```

#### Interval halving method

```python

a1=w1
b1=w3
print(a1,b1)
l=b1-a1

e1=0.00000000000001

while abs(l)>=e1:
    
    wm=(a1+b1)/2
    x1=a1+l/4
    x2=b1-l/4

    if J(x1)<J(wm):
        b1=wm
        wm=x1
           
    elif J(x2)<J(wm):
        a1=wm
        wm=x2
    
    else:
        a1=x1
        b1=x2
    
    l=b1-a1
    range2=(a1,b1)
print('The critical point lies in the the range:',range2)

```

#### Newton Raphson method

```python


def dJ(w):
    if w==0:
        return float('inf')
    else:
        return 2*w-54/w**2
def d2J(w):
    if w==0:
        return float('inf')
    else:
        return 2+108/w**3

p1=1
e2=0.00000000001
k=1

while abs(dJ(p1))>=e2:
    p2= p1-dJ(p1)/d2J(p1)
    k=k+1
    p1=p2
    
print('The critical point is:',p1)
print('The number of iterations to reach this value are:',k)

```
