```python
import matplotlib.pyplot as plt
import numpy as np

class Matrix:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    
    def elementwise_addition(self):
        if self.x.shape==self.y.shape:
            self.z=np.add(self.x,self.y)
            print('The elementwise addition is:',self.z)
        else:
            raise ValueError('Value error: Number of rows and columns of the matrices should be equal')
    
    def elementwise_subtraction(self):
        if self.x.shape==self.y.shape:
            self.z=np.subtract(self.x,self.y)
            print('The elementwise subtraction is:',self.z)
        else:
            raise ValueError('Value error: Number of rows and columns of the matrices should be equal')
        
    def elementwise_multiplication(self):
        if self.x.shape==self.y.shape:
            self.z=np.multiply(self.x,self.y)
            print('The elementwise multiplication is:',self.z)
        else:
            raise ValueError('Value error: Number of columns of the first matrix should be equal to the number of rows of the second matrix')
    
    def scalar_multiplication(self,k):
        print('The scalar multiplication of x with k is:',self.x*k)
        print('The scalar multiplication of y with k is:',self.y*k)
    
    def determinant(self):
        if self.x.shape[0]==self.x.shape[1] and self.y.shape[0]==self.y.shape[1]:
            self.u=np.linalg.det(self.x)
            self.v=np.linalg.det(self.y)
            print('The determinant of x is ',round(self.u,2))
            print('The determinant of y is ',round(self.v,2))
        else:
            raise ValueError('The matrices should be square for calculating determinant')
    def transpose(self):
        print('The transpose of x is:',np.transpose(self.x))
        print('The transpose of y is:',np.transpose(self.y))
    def inverse(self):
        if self.u!=0 and self.v!=0:
            print('The inverse of x is :',np.linalg.inv(self.x))
            print('The inverse of y is :',np.linalg.inv(self.y))
        else:
             raise ValueError('The determinant of a matrix should not be 0 for calculating inverse')
    def diagonal(self):
        if self.x.shape[0]==self.x.shape[1]:
            print('The diagonal elements of the first matrix are:',self.x.diagonal())
        else:
            raise ValueError('The matrix should be a square matrix')
    def rank(self):
        print('The rank of matrix x is:',np.linalg.matrix_rank(self.x))
        print('The rank of matrix y is:',np.linalg.matrix_rank(self.y))
    def eigen(self):
        if self.x.shape[0]==self.x.shape[1]:
            u,v=np.linalg.eig(self.x)
            print('The eigenvalues of the first matrix are:',u)
            print('The eigenvectors of the first matrix are',v)
            a=[v[0][0],0]
            b=[v[0][1],0]
            c=[v[1][0],0]
            d=[v[1][1],0]
            plt.plot(a,b)
            plt.plot(c,d)

        else:
            raise ValueError('The matrices should be square for calculating eigenvalues and eigen vectors')

        
            
obj1=Matrix(np.array([[5, 10,1], [15, 20,45]]),np.array([[25, 30,2], [35, 40,5]]))
obj2=Matrix(np.array([[1,2],[2,3]]),np.array([[2,6],[7,8]]))
obj1.elementwise_addition()
obj1.elementwise_subtraction()
obj2.elementwise_multiplication()
obj1.scalar_multiplication(4)
obj2.determinant()
obj1.transpose()
obj2.inverse()
obj2.diagonal()
obj1.rank()
obj2.eigen()




     
    
        
```
